```python
import numpy as np
import tensorflow as tf
```

```python
x=np.arange(-10,10,0.1).astype(np.float32)
y=tf.nn.sigmoid(x)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
y_=sess.run(y)

import matplotlib.pyplot as plt
plt.plot(x,y_)
plt.show()
```

```python
x=np.arange(-10,10,0.1).astype(np.float32)
y=tf.nn.relu(x)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
y_=sess.run(y)

import matplotlib.pyplot as plt
plt.plot(x,y_)
plt.show()
```

```python
x=np.arange(-10,10,0.1).astype(np.float32)
y=tf.nn.relu6(x)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
y_=sess.run(y)

import matplotlib.pyplot as plt
plt.plot(x,y_)
plt.show()
```
